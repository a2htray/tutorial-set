```python
import squarify
import matplotlib.pyplot as plt
```

```python
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
```

```python
squarify.plot(
    sizes=[20, 10, 30, 40], 
    label=['组 ' + str(i) for i in range(1, 5)], 
    color=['red', 'green', 'blue', 'grey'],
    alpha=.4,
)
plt.axis('off')
plt.title('不同组用户比例')
plt.show()
```
