```python
import matplotlib.pyplot as plt
```

```python
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
```

```python
x = range(21, 26)
y = [
    [10, 4, 6, 5, 3],
    [12, 2, 7, 10, 1],
    [8, 18, 5, 7, 6],
    [1, 8, 3, 5, 9],
]
labels = ['组 ' + str(i) for i in range(1, 5)]
```

```python
plt.stackplot(x, y, labels=labels, alpha=.7)
plt.ylabel('分值')
plt.xlabel('年龄')
plt.title('不同组用户区间分值比较')
plt.legend(loc='upper right')
```
