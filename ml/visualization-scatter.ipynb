```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
```

```python
np.random.seed(42)

n = 40
x = np.random.random(n)
y = np.random.random(n)
colors = np.random.random(n)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(8, 4))

ax1.scatter(x, y, s=100, c='black', alpha=.8)
ax1.set_xlabel('X 坐标')
ax1.set_ylabel('Y 坐标')
ax1.set_title('随机生成的数字散点')

ax2.scatter(x, y, c=colors, s=y*1000, alpha=0.9)
ax2.set_xlabel('X 坐标')
ax2.set_ylabel('Y 坐标')
ax2.set_title('随机生成的数字散点')
```
