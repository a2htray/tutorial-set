```python
import matplotlib.pyplot as plt
```

```python
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
```

```python
labels = ['财经 15%', '社会 30%', '体育 15%', '科技 10%', '其他 30%']
sizes = [15, 30, 15, 10, 30]
explode = [0, 0.1, 0, 0, 0]

fig, ax = plt.subplots()
pie = ax.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', shadow=False, startangle=90)

for patch, hatch in zip(pie[0], ['.', '-', '+', 'X', 'o']):
    patch.set_hatch(hatch)

fig.legend()
ax.axis('equal')
plt.title('新闻网站用户兴趣分析')
```
