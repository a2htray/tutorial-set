```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Circle
```

```python
# 圆
fig, ax = plt.subplots(2, 2)

# 生成一个 (2, 2) 坐标上半径为 2、填充颜色为 white 和轮廓颜色为 cornflowerblue 的圆
# 由于坐标轴的刻度线的变化量不一致，所以显示的圆是一个椭圆。
circle = Circle((2, 2), radius=2, facecolor='white', edgecolor='cornflowerblue')
# 将圆添加到坐标系中
ax[0,0].add_patch(circle)
ax[0,0].set_xlim(-1, 5)
ax[0,0].set_ylim(-1, 5)

# 获取坐标系的矩形区域
rectangle = ax[0,1].patch
rectangle.set_facecolor('gold')
circle = Circle((2, 2), radius=2, facecolor='white', edgecolor='cornflowerblue')
ax[0,1].add_patch(circle)
ax[0,1].set_xlim(-1, 5)
ax[0,1].set_ylim(-1, 5)
# equal 表示 aspect 为 1，即 X 轴与 Y 轴显示一样
# box 表示按坐标轴长度 set_aspect
ax[0,1].set_aspect('equal', 'box')

rectangle = ax[1,0].patch
rectangle.set_facecolor('palegreen')
circle = Circle((2, 2), radius=2, facecolor='white', edgecolor='cornflowerblue')
ax[1,0].add_patch(circle)
# equal 通过修改坐标轴限制对坐标进行伸缩
ax[1,0].axis('equal')

rectangle = ax[1,1].patch
rectangle.set_facecolor('lightskyblue')
circle = Circle((2, 2), radius=2, facecolor='white', edgecolor='cornflowerblue')
ax[1,1].add_patch(circle)
# 统一设置坐标轴的显示区间
ax[1,1].axis([-1, 6, -1, 6])
ax[1,1].set_yticks(np.arange(-1, 6, 1))
# ax[1,1].axis('equal')

plt.subplots_adjust(left=0.1)
```

```python
from matplotlib.patches import Ellipse
```

```python
# 椭圆
fig, (ax1, ax2) = plt.subplots(1, 2, subplot_kw={'aspect': 'equal'})

# 生成 4 个旋转角度
angles = np.linspace(0, 135, 4)
ellipses = [
    Ellipse(
        xy=(2, 2), 
        width=4, 
        height=2, 
        angle=a,
    ) 
    for a in angles
]
for ellipse in ellipses:
    ax1.add_patch(ellipse)
    # 设置椭圆的透明度
    ellipse.set_alpha(0.4)
    # 设置椭圆填充颜色
    ellipse.set_color('cornflowerblue')

ax1.axis([-1, 5, -1, 5])

ellipses = [
    Ellipse(
        xy=np.random.rand(2) * 10,
        width=np.random.rand(1),
        height=np.random.rand(1),
        angle=np.random.rand(1) * 360,
    )
    for _ in range(0, 100, 1)
]

for ellipse in ellipses:
    ax2.add_patch(ellipse)
    ellipse.set_alpha(np.random.random())
    ellipse.set_color(np.random.rand(3))
    
ax2.axis([-1, 11, -1, 11])

plt.tight_layout()
```
