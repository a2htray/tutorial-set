```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 规则多边形的颜色填充
x = [0, 0, 5, 10, 15, 15, 10, 5]
y = [5, 10, 15, 15, 10, 5, 0, 0]

# x y 是有序数对，用 plt.fill 方法完成对规则图形的颜色填充
plt.fill(x, y, color='cornflowerblue')
# 绘制几何图形顶点
plt.scatter(x, y, c='red', s=200)

plt.xlim(-1, 16)
plt.ylim(-1, 16)
plt.xticks(np.arange(0, 16, 5))
plt.yticks(np.arange(0, 16, 5))
```

```python
x = np.linspace(0, 2 * np.pi, 500)
y = np.sin(x)

# 不规则多边形的颜色填充
plt.fill(x, y, color='cornflowerblue', alpha=0.4)
# 描边
plt.plot(x, y, color='cornflowerblue', alpha=0.8)
# 曲线首尾坐标描色
plt.plot([x[0], x[-1]], [y[0], y[-1]], color='cornflowerblue', alpha=0.8)

plt.xlim(0, 2 * np.pi)
plt.ylim(-1.1, 1.1)
```

```python
# 交叉曲线的颜色填充
x = np.linspace(0, 2, 500)
y1 = np.sin(2 * np.pi * x)
y2 = 1.1 * np.sin(3 * np.pi * x)

# sharex=all 共享 X 轴
fig, (ax1, ax2, ax3, ax4) = plt.subplots(4, 1, sharex='all')
ax1.plot(x, y1, color='red')
ax1.plot(x, y2, color='blue')
ax1.set_ylim(-1.2, 1.2)

ax2.fill_between(x, 0, y2, color='#9c9c9c', alpha=0.5)
ax2.set_ylim(-1.2, 1.2)

ax3.fill_between(x, y2, 1.1, color='#9c9c9c', alpha=0.5)
ax3.set_ylim(-1.2, 1.2)

ax4.fill_between(x, y1, y2, color='#9c9c9c', alpha=0.5)
ax4.set_xlim(0, 2)
ax4.set_ylim(-1.2, 1.2)
```

```python
# 使用 fill_between 和 fill_betweenx 实现水平和垂直交叉曲线封闭区域的颜色填充
fig, (ax1, ax2) = plt.subplots(1, 2)

x = np.linspace(0, 2, 500)
y1 = np.sin(2 * np.pi * x)
y2 = 1.1 * np.sin(3 * np.pi * x)

ax1.plot(x, y1, color='k', lw=1, ls='-')
ax1.plot(x, y2, color='k', lw=1, ls='-')
# 在 y2 >= y1 的封闭区域上填充 cornflowerblue
ax1.fill_between(x, y1, y2, where=y2>=y1, interpolate=True, facecolor='cornflowerblue', alpha=0.7)
# 在 y2 <= y1 的封闭区域上填充 darkred
ax1.fill_between(x, y1, y2, where=y2<=y1, interpolate=True, facecolor='darkred', alpha=0.7)
# 绘制网格
ax1.grid(ls=':', lw=1, color='gray', alpha=0.5)

y = np.linspace(0, 2, 500)
x1 = np.sin(2 * np.pi * y)
x2 = 1.1 * np.sin(3 * np.pi * y)

ax2.plot(x1, y, color='k', lw=1, ls='-')
ax2.plot(x1, y, color='k', lw=1, ls='-')
# 在 x1 >= x2 的封闭区域上填充 cornflowerblue
ax2.fill_betweenx(y, x1, x2, where=x1>=x2, interpolate=True, facecolor='cornflowerblue', alpha=0.7)
# 在 x1 <= x2 的封闭区域上填充 gray
ax2.fill_betweenx(y, x1, x2, where=x1<=x2, interpolate=True, facecolor='darkred', alpha=0.7)
# 绘制网格
ax2.grid(ls=':', lw=1, color='gray', alpha=0.5)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2)
x = np.linspace(0, 2, 500)
y1 = np.sin(2 * np.pi * x)
y2 = 1.2 * np.sin(3 * np.pi * x)
# 对 y2 中大于 1.0 的值进行掩饰，返回一个 masked_array(data, mask)
# 如果 mask 数组中的值为 True 时，则不在图像中显示
y2 = np.ma.masked_greater(y2, 1.0)

ax1.plot(x, y1, color='k', lw=1, ls='-')
ax1.plot(x, y2, color='k', lw=1, ls='-')
ax1.fill_between(x, y1, y2, where=y2>=y1, facecolor='cornflowerblue', alpha=0.7)
ax1.fill_between(x, y1, y2, where=y1>=y2, facecolor='darkred', alpha=0.7)
ax1.set_xlim(0, 2)
ax1.set_ylim(-1.2, 1.2)
ax1.grid(ls=':', lw=1, color='gray', alpha=0.5)

y = np.linspace(0, 2)
x1 = np.sin(2 * np.pi * y)
x2 = 1.2 * np.sin(3 * np.pi * y)
x2 = np.ma.masked_greater(x2, 1.0)

ax2.plot(x1, y, color='k', lw=1, ls='-')
ax2.plot(x2, y, color='k', lw=1, ls='-')
ax2.fill_betweenx(y, x1, x2, where=x2>=x1, facecolor='cornflowerblue', alpha=0.7)
ax2.fill_betweenx(y, x1, x2, where=x1>=x2, facecolor='darkred', alpha=0.7)
ax2.set_xlim(-1.2, 1.2)
ax2.set_ylim(0, 2)
ax2.grid(ls=':', lw=1, color='gray', alpha=0.5)
```
